```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# --- Load and Preprocess the Data ---

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target labels

# Standardize the data (PCA works better with standardized data)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# --- PCA without Whitening ---

# Apply PCA
pca = PCA(n_components=2, whiten=False)
X_pca = pca.fit_transform(X_scaled)

# Plot the principal components (without whitening)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.title('PCA without Whitening')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class Labels')
plt.grid(True)
plt.show()

# --- PCA with Whitening ---

# Apply PCA with Whitening
pca_whitened = PCA(n_components=2, whiten=True)
X_pca_whitened = pca_whitened.fit_transform(X_scaled)

# Plot the principal components (with whitening)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca_whitened[:, 0], X_pca_whitened[:, 1], c=y, cmap='plasma', edgecolor='k', s=100)
plt.title('PCA with Whitening')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class Labels')
plt.grid(True)
plt.show()
```

```python
# --- Explained Variance ---

# Explained variance ratio without whitening
explained_variance = pca.explained_variance_ratio_
print(f'Explained Variance (without Whitening): {explained_variance}')

# Explained variance ratio with whitening
explained_variance_whitened = pca_whitened.explained_variance_ratio_
print(f'Explained Variance (with Whitening): {explained_variance_whitened}')
```
